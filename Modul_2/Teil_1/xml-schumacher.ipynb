## Bereinigte bibsonomy-Daten in eine xml-Datei wandeln

```python
import json
```

```python
with open("bibsonomy_fields_deleted_names_changed.json", "r") as file:
    xml_data = json.load(file)
```

```python
print(len(xml_data))
```

```python
print(xml_data)
```

```python
print("<?xml version=\"1.0\" encoding=\"UTF-8\"?>")
print("<publications>")
   
for publication in xml_data:
    print("\t<publication>")   

            
    for key in publication.keys():
        if key == "tags":                                   
            for tag in publication["tags"]:
                print("\t\t" + "<tags>" + str(tag) + "</tags>")
                
        elif key == "author":
            for author in publication["author"]:
                print("\t\t" + "<author>" + str(author) + "</author>")
    
        elif key == "editor":
            for editor in publication["editor"]:
                print("\t\t" + "<editor>" + str(editor) + "</editor>")
    
        elif key == "authors":
            for authors in publication["authors"]:
                if type(authors) is dict:
                    for name in authors.keys():
                        print("\t\t" + "<" + str(key) + ">")
                        print("\t\t\t" + "<" + str(name) + ">" + str(authors[name]) + "</" + str(name) +">")
                        print("\t\t" + "</" + str(key) + ">")
  
        elif key == "editors":
            for editors in publication["editors"]:
                if type(editors) is dict:
                    for name_e in editors.keys():
                        print("\t\t" + "<" + str(key) + ">")
                        print("\t\t\t" + "<" + str(name_e) + ">" + str(authors[name_e]) + "</" + str(name_e) +">")
                        print("\t\t" + "</" + str(key) + ">")


        else:
            print("\t\t" + "<" + str(key) + ">" + str(publication[key]) + "</" + str(key) + ">")
                        
    print("\t</publication>")
print("</publications>")  
    
```

```python
with open ("xml-final.xml", "w") as xml:
    

    print("<?xml version=\"1.0\" encoding=\"UTF-8\"?>", file=xml)
    print("<publications>", file=xml)
   
    for publication in xml_data:
        print("\t<publication>", file=xml)   

            
        for key in publication.keys():
            if key == "tags":                                   
                for tag in publication["tags"]:
                    print("\t\t" + "<tags>" + str(tag) + "</tags>", file=xml)
                
            elif key == "author":
                for author in publication["author"]:
                    print("\t\t" + "<author>" + str(author) + "</author>", file=xml)
        
            elif key == "editor":
                for editor in publication["editor"]:
                    print("\t\t" + "<editor>" + str(editor) + "</editor>", file=xml)
        
        
            elif key == "authors":
                for authors in publication["authors"]:
                    if type(authors) is dict:
                        for name in authors.keys():
                            print("\t\t" + "<" + str(key) + ">", file=xml)
                            print("\t\t\t" + "<" + str(name) + ">" + str(authors[name]) + "</" + str(name) +">", file=xml)
                            print("\t\t" + "</" + str(key) + ">", file=xml)
  
            elif key == "editors":
                for editors in publication["editors"]:
                    if type(editors) is dict:
                        for name_e in editors.keys():
                            print("\t\t" + "<" + str(key) + ">", file=xml)
                            print("\t\t\t" + "<" + str(name_e) + ">" + str(authors[name_e]) + "</" + str(name_e) +">", file=xml)
                            print("\t\t" + "</" + str(key) + ">", file=xml)


            else:
                print("\t\t" + "<" + str(key) + ">" + str(publication[key]) + "</" + str(key) + ">", file=xml)
                        
        print("\t</publication>", file=xml)
    print("</publications>", file=xml)  
    
```

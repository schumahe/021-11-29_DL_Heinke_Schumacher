```python
import json
```

```python
with open("pubs_books_bibsonomy.json", "r") as file:
    data = json.load(file)
```

```python
print(len(data))
```

```python
print(json.dumps(data, indent=4))
```

```python
for item in data:
    for key in item.keys():
        print(key)
```

```python
edited_items = []
```

```python
desired_items = ["type", "id", "tags", "description", "label", "date", "changeDate", "count", "pub-type", "publisher",
                 "year", "url", "author", "authors", "abstract", "isbn", "language"]
```

```python
for item in data:
    item_export = {}
    for entry in desired_items:
        if entry in item.keys():
            item_export[entry] = item[entry]

    edited_items.append(item_export)
```

```python
print(edited_items)
```

```python
print(json.dumps(edited_items, indent=4))
```

```python
with open ("new.xml", "w") as xml:
    print("<?xml version=\"1.0\" encoding=\"UTF-8\"?>", file=xml)
    print("<publications>", file=xml)
    for publication in edited_items:
        print("\t<publication>", file=xml)
        for key in publication.keys():
              print("\t\t" + "<" + str(key) + ">" + str(publication[key]) + "<\\" + str(key) + ">", file=xml)
    print("<\\publications>\n", file=xml)
```

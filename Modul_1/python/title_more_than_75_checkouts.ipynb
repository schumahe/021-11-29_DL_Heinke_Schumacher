## Für welche Titel und Verlage betragen die Ausleihen mehr als 75?

```python
import urllib.request
import pandas as pd
```

```python
url = "https://data.seattle.gov/resource/tmmm-ytt6.csv?$where=checkoutyear=2020&$limit=10000"
checkout_file = "checkout.csv"
urllib.request.urlretrieve(url, checkout_file)
```

```python
checkouts_df = pd.read_csv(checkout_file)
```

```python
checkouts_75_df = checkouts_df[checkouts_df["checkouts"] >= 75]
```

```python
checkouts_75_df [["checkouts", "title", "publisher"]]
```

```python
checkouts_75_df ["checkouts"].hist()
```

```python
checkouts_75_df.sort_values(by=["checkouts"]) [["checkouts", "title", "publisher"]]
```

```python
checkouts_75sort_df = (checkouts_75_df.sort_values(by=["checkouts"]) [["checkouts", "title", "publisher"]])
```

```python
checkouts_75sort_df.plot(kind="bar", x="title", y="checkouts", fontsize=10)
```

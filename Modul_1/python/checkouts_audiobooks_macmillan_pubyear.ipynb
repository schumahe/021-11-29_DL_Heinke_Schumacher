## Wieviele Checkouts von Audiobooks für den Verlag Macmillan?

```python
import urllib.request
import pandas as pd
```

```python
url = "https://data.seattle.gov/resource/tmmm-ytt6.csv?$where=checkoutyear=2020&$limit=10000"
checkout_file = "checkout.csv"
urllib.request.urlretrieve(url, checkout_file)
```

```python
checkouts_df = pd.read_csv(checkout_file)
```

```python
checkouts_df
```

```python
checkouts_df.materialtype.unique()
```

```python
checkouts_df[checkouts_df["materialtype"] == "AUDIOBOOK"]
```

```python
checkouts_audiobooks_df = checkouts_df[checkouts_df["materialtype"] == "AUDIOBOOK"]
```

```python
checkouts_audiobooks_df [["checkouts", "publisher", "materialtype"]]
```

```python
checkouts_audiobooks_df["publisher"].value_counts()
```

```python
checkouts_audiobooks_df[checkouts_audiobooks_df["publisher"] == "Macmillan Audio"]
```

```python
checkouts_audiobooks_macmillan_df = checkouts_audiobooks_df[checkouts_audiobooks_df["publisher"] == "Macmillan Audio"]
```

```python
checkouts_audiobooks_macmillan_df.sort_values(by="publicationyear")[["checkouts", "checkoutmonth", "publicationyear"]]
```

```python
checkouts_audiobooks_macmillan_df.groupby("publicationyear").agg({"checkouts": ["sum"]})
```

```python
checkouts_audiobooks_pubyear_macmillan_df = checkouts_audiobooks_macmillan_df.groupby("publicationyear").agg({"checkouts": ["sum"]})
```

```python
%matplotlib inline
```

```python
checkouts_audiobooks_pubyear_macmillan_df.plot(kind="bar")
```

```python
checkouts_audiobooks_pubyear_macmillan_df.plot(kind="pie", y="checkouts", fontsize=12, figsize=(10,10))
```
